```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = './data/mouse_drug_data.csv'
clinical_trial_data_to_load = './data/clinicaltrial_data.csv'

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_data = pd.read_csv("data/mouse_drug_data.csv")
clinical_trial_data = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
combined_data = pd.merge(mouse_drug_data, clinical_trial_data, on="Mouse ID", how = "left")

# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_volume = combined_data.groupby(["Drug", "Timepoint"])
mean_tumor = tumor_volume["Tumor Volume (mm3)"].mean()
mean_tumor

# Convert to DataFrame
tumor_response = pd.DataFrame(mean_tumor).reset_index()

# Preview DataFrame
tumor_response.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_volume_error = tumor_volume["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
tumor_response_error = pd.DataFrame(tumor_volume_error).reset_index()

# Preview DataFrame
tumor_response_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
data_munging = tumor_response.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
data_munging
```

```python
#Generate the Plot (with Error Bars)
capomulin_error = tumor_response_error.loc[tumor_response_error["Drug"] == "Capomulin", "Tumor Volume (mm3)"]
infubinol_error = tumor_response_error.loc[tumor_response_error["Drug"] == "Infubinol", "Tumor Volume (mm3)"]
ketapril_error = tumor_response_error.loc[tumor_response_error["Drug"] == "Ketapril", "Tumor Volume (mm3)"]
placebo_error = tumor_response_error.loc[tumor_response_error["Drug"] == "Placebo", "Tumor Volume (mm3)"]
time = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]

plt.errorbar(data_munging.index, data_munging["Capomulin"] , yerr= capomulin_error, label= "Capomulin", marker= "o", color="red", linestyle='--')
plt.errorbar(data_munging.index, data_munging["Infubinol"] , yerr= infubinol_error, label= "Infubinol", marker= "^", color="blue", linestyle='--')
plt.errorbar(data_munging.index, data_munging["Ketapril"] , yerr= ketapril_error, label= "Ketapril", marker= "D", color="black", linestyle='--')
plt.errorbar(data_munging.index, data_munging["Placebo"] , yerr= placebo_error , label= "Placebo", marker= "s", color="green", linestyle='--')
plt.legend()
plt.title("Tumor Response to Treatment ")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
# Save the Figure
#plt.savefig("Images/Tumor_Response.png")
```
